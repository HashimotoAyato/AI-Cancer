```python
from sklearn import metrics
import matplotlib.pyplot as plt
import csv
import os
```

```python
LOAD_DIR = '/export/hashimoto/Results'
LOAD_PROJECT = 'A6c'
LOAD_FILE = 'epoch79_valid_output_label.csv'
SAVE_DIR = '/export/hashimoto/ROC/'
```

```python
csv_file = open(os.path.join(LOAD_DIR,LOAD_PROJECT,LOAD_FILE), 'r')
csv_list = list(csv.reader(csv_file, delimiter=','))
pred, ans = [], []
for data in csv_list:
    pred.append(float(data[0]))
    ans.append(float(data[1]))
```

```python
fpr, tpr, thresholds = metrics.roc_curve(ans, pred)
auc = metrics.auc(fpr, tpr)
plt.figure(facecolor='white', edgecolor='black')
plt.plot(fpr, tpr, marker='o', markersize=1)
plt.xlabel('FPR: False positive rate')
plt.ylabel('TPR: True positive rate')
plt.grid()
plt.savefig(SAVE_DIR + LOAD_PROJECT + '_' + LOAD_FILE + '_roc.png')
print('AUC:{}'.format(auc))
```
